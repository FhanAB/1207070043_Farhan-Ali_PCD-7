Import Library

```python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

Gray Level Slice

```python
img = data.horse()  # Mengambil citra "camera" dari modul "data" dalam library "skimage" dan menyimpannya ke dalam variabel "img".

row, column = img.shape  # Mengambil dimensi baris dan kolom dari citra "img" dan menyimpannya ke dalam variabel "row" dan "column" secara terpisah.

img1 = np.zeros((row,column),dtype = 'uint8')  
# Membuat array numpy dengan ukuran yang sama dengan citra "img" berdasarkan dimensi baris dan kolomnya.
# Array ini diinisialisasi dengan nilai 0 dan tipe data "uint8" (unsigned integer 8-bit), kemudian disimpan dalam variabel "img1".

min_range = 10  # Mendefinisikan nilai ambang bawah "min_range" untuk proses thresholding.
max_range = 60  # Mendefinisikan nilai ambang atas "max_range" untuk proses thresholding.

for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:  # Jika nilai piksel lebih besar dari "min_range" dan lebih kecil dari "max_range",
            img1[i,j] = 255  # piksel pada citra "img1" diubah menjadi 255 (putih).
        else:
            img1[i,j] = 0  # Jika tidak memenuhi kondisi tersebut, piksel pada citra "img1" diubah menjadi 0 (hitam).
```

Plot Image

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))# Membuat grid subplot dengan 2 baris dan 2 kolom, dengan ukuran total 12x12, dan menyimpan objek subplot dan figurenya dalam variabel "axes" dan "fig".
ax = axes.ravel() # Mengubah array objek subplot menjadi array 1 dimensi menggunakan "ravel()" dan menyimpannya dalam variabel "ax".


ax[0].imshow(img, cmap=plt.cm.gray)# Menampilkan citra "img" pada subplot pertama
ax[0].set_title("Citra Input")# Memberikan judul pada subplot pertama.

ax[1].hist(img.ravel().astype(float), bins=256)# Menampilkan histogram dari citra "img" pada subplot kedua
ax[1].set_title('Histogram Input')# Memberikan judul pada subplot kedua.

ax[2].imshow(img1, cmap=plt.cm.gray) # Menampilkan citra "img1" pada subplot ketiga.
ax[2].set_title("Citra Output")# Memberikan judul pada subplot ketiga.

ax[3].hist(img1.ravel().astype(float), bins=256)# Menampilkan histogram dari citra "img1" pada subplot keempat
ax[3].set_title('Histogram Output')# Memberikan judul pada subplot keempat.

# Mengatur tata letak (layout) subplot agar tidak tumpang tindih.
fig.tight_layout()
```
