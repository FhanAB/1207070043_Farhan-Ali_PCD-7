Filter Median 1

```python
for baris in range(0, m1-1):  # Iterasi melalui setiap baris citra 1
    for kolom in range(0, n1-1):  # Iterasi melalui setiap kolom citra 1
        a1 = baris  # Koordinat baris saat ini
        b1 = kolom  # Koordinat kolom saat ini
        b1 = kolom
         # Mengumpulkan data sekitar (3x3) dari citra 1
        dataA = [copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1-1, b1+1], \
              copyCitra1[a1, b1-1], copyCitra1[a1, b1], copyCitra1[a1, b1+1], \
              copyCitra1[a1+1, b1-1], copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]]
        
        # Urutkan data secara ascending
        for i in range(1, 8):  # Iterasi melalui indeks 1 hingga 7
            for j in range(i, 9):  # Iterasi melalui indeks i hingga 8
                if dataA[i] > dataA[j]:  # Periksa apakah elemen pada indeks i lebih besar dari elemen pada indeks j
                  tmpA = dataA[i]  # Simpan nilai elemen pada indeks i ke dalam variabel sementara
                  dataA[i] = dataA[j]  # Gantikan nilai elemen pada indeks i dengan nilai elemen pada indeks j
                  dataA[j] = tmpA  # Gantikan nilai elemen pada indeks j dengan nilai sementara yang disimpan sebelumnya
        
        output1[a1, b1] = dataA[5]  # Ambil data median dan simpan ke dalam output citra 1
```

Filter Median 2

```python
for baris in range(0, m2-1):  # Iterasi melalui setiap baris citra
    for kolom in range(0, n2-1):  # Iterasi melalui setiap kolom citra
        a1 = baris  # Simpan nilai baris saat ini dalam variabel a1
        b1 = kolom  # Simpan nilai kolom saat ini dalam variabel b1
        dataA = [copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1-1, b1+1], \
              copyCitra2[a1, b1-1], copyCitra2[a1, b1], copyCitra2[a1, b1+1], \
              copyCitra2[a1+1, b1-1], copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]]
        # Membentuk list dataA yang berisi nilai piksel di sekitar piksel saat ini
        
        # Urutkan
        for i in range(1, 8):  # Iterasi melalui indeks 1 hingga 7
            for j in range(i, 9):  # Iterasi melalui indeks i hingga 8
                if dataA[i] > dataA[j]:  # Periksa apakah elemen pada indeks i lebih besar dari elemen pada indeks j
                    tmpA = dataA[i]  # Simpan nilai elemen pada indeks i ke dalam variabel sementara
                    dataA[i] = dataA[j]  # Gantikan nilai elemen pada indeks i dengan nilai elemen pada indeks j
                    dataA[j] = tmpA  # Gantikan nilai elemen pada indeks j dengan nilai sementara yang disimpan sebelumnya
        
        output2[a1, b1] = dataA[5]  # Simpan nilai piksel terurut di output2 pada posisi baris dan kolom saat ini
```
